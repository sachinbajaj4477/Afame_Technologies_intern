```python
# 1. Removing unnecessary columns
unnecessary_columns = ['EmployeeCount', 'Over18', 'StandardHours']  # Columns to remove
df_cleaned = df.drop(columns=unnecessary_columns)
```

```python

# 2. Giving the columns new names (if needed)
# Example: Let's rename some columns for better clarity
column_names_mapping = {
    'Age': 'Age',
    'Attrition': 'Attrition',
    'BusinessTravel': 'Business_Travel',
    # Add more columns as needed
}
df_cleaned = df_cleaned.rename(columns=column_names_mapping)
```

```python
# 3. Eliminating redundant entries
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# 4. Sanitizing specific columns
for column in df_cleaned.columns:
    # Check if the column is of string type
    if df_cleaned[column].dtype == 'object':
        # Remove leading and trailing whitespaces from column values
        df_cleaned[column] = df_cleaned[column].str.strip()
```

```python
# 5. Eliminating NaN values
df_cleaned = df_cleaned.dropna()
```

```python

# Display the cleaned dataset
print("\nCleaned Dataset:")
df_cleaned.head()
```

```python
# 1. Histogram of Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df_cleaned['Age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Bar plot of Attrition
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=df_cleaned, palette='Set2')
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Box plot of Monthly Income by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='MonthlyIncome', data=df_cleaned, palette='Set3')
plt.title('Monthly Income by Department')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Check for missing values in 'BusinessTravel' column
print("Missing values in 'BusinessTravel' column:", df_cleaned['Business_Travel'].isnull().sum())

# Check unique values in 'BusinessTravel' column
print("Unique values in 'BusinessTravel' column:", df_cleaned['Business_Travel'].unique())
```

```python
# Visualize all columns
for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':
        # Categorical column: create count plot
        plt.figure(figsize=(8, 6))
        sns.countplot(x=column, data=df_cleaned, palette='Set2')
        plt.title(f'{column} Count')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.show()
    else:
        # Numerical column: create histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], bins=30, kde=True, color='skyblue')
        plt.title(f'{column} Distribution')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```
